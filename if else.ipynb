```python
name = input("Enter your Name: ")
subject1 = int(input("Enter subject 1 (out of 100): "))
subject2 = int(input("Enter subject 2 (out of 100): "))
subject3 = int(input("Enter subject 3 (out of 100): "))
subject4 = int(input("Enter subject 4(out of 100): "))
subject5 = int(input("Enter subject 5 (out of 100): "))
total_marks = 500
obt_marks = int(subject1 + subject2 + subject3 + subject4 + subject5)
# print(obt_marks)
percentage = float(obt_marks / total_marks * 100)
# print(percentage)
if percentage >= 90:
    grade = "A"
elif percentage >= 80:
    grade = "B"
elif percentage >= 70:
    grade = "C"
elif percentage >= 60:
    grade = "D"
elif percentage >= 50:
    grade = "E"
else:
    grade = "F"
print(f"Your Class Report        \n Name: {name}\n obtained Marks: {obt_marks}\n percentage: {percentage}%\n Grade; {grade}")
```
