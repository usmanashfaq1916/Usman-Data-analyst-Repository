```python
# Initialize an empty list to store the marks
marks_list = []

# Loop to input marks for 5 students
for i in range(1, 6):
    while True:
        try:
            # Take input and convert to floating-point number
            mark = float(input(f"Enter marks for Student {i}: "))
            if 0 <= mark <= 100:  # Ensures valid exam score range
                marks_list.append(mark)
                break
            else:
                print("Please enter a valid mark between 0 and 100.")
        except ValueError:
            print("Invalid input. Please enter a numerical value.")

# Calculate statistics
highest_marks = max(marks_list)
lowest_marks = min(marks_list)
average_marks = sum(marks_list) / len(marks_list)

# Count how many students passed (marks >= 40)
passed_students = sum(1 for mark in marks_list if mark >= 40)

# Print the results
print("\n--- Student Performance Report ---")
print(f"Highest marks: {highest_marks}")
print(f"Lowest marks: {lowest_marks}")
print(f"Average marks: {average_marks:.2f}")
print(f"Number of students who passed (>=40): {passed_students}")
```

```python
"""
Banking System
"""

userDb = {
    "name": [],
    "balance": []
}

while True:

    userChoice = int(input(
        "\nEnter 1 to Create Account"
        "\nEnter 2 to Deposit, Withdraw or Check Balance"
        "\nEnter 3 to Exit\n"
    ))

    if userChoice == 1:

        userName = input("Enter Account Holder Name: ")

        if userName not in userDb["name"]:
            userDb["name"].append(userName)
            userDb["balance"].append(500)
            print(f"Account Created Successfully.")
            print("Opening Balance = 500")

        else:
            print("User already exists.")

    elif userChoice == 2:

        name = input("Enter Account Holder Name: ")

        if name in userDb["name"]:

            index = userDb["name"].index(name)

            preference = int(input(
                "\n1. Deposit"
                "\n2. Withdraw"
                "\n3. Check Balance\n"
            ))

            if preference == 1:

                deposit = float(input("Enter Deposit Amount: "))
                userDb["balance"][index] += deposit

                print("Deposit Successful")
                print("Current Balance:", userDb["balance"][index])

            elif preference == 2:

                withdraw = float(input("Enter Withdrawal Amount: "))

                if withdraw <= userDb["balance"][index]:

                    userDb["balance"][index] -= withdraw

                    print("Withdrawal Successful")
                    print("Current Balance:", userDb["balance"][index])

                else:
                    print("Insufficient Balance")

            elif preference == 3:

                print("Available Balance:", userDb["balance"][index])

            else:
                print("Invalid Choice")

        else:
            print("Account does not exist.")

    elif userChoice == 3:

        print("Exit Successfully")
        break

    else:
        print("Invalid Choice")
```
