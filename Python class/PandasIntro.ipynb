```python
quec
score = 0

for question, data in quec.items():
    print("\n" + question + "\n")

    for option in ["A", "B", "C", "D"]:
        print(option + ".", data[option])

    user = input("Enter your answer (A/B/C/D): ").upper()

    if user == data["answer"]:
        print("✅ Correct!")
        score += 1
    else:
        print("❌ Wrong!")
        print("Correct Answer:", data["answer"])

print("\nYour Score:", score, "/", len(quiz))
```
