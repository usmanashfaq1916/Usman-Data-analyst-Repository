# File Handling in Python

Python can read from and write to files on disk. The `open()` function is used to open a file and returns a file object. The `with` statement ensures the file is properly closed after use.

**File modes:**
- `"r"` — read (default)
- `"w"` — write (overwrites existing file)
- `"a"` — append (adds to end)
- `"r+"` — read + write
- `"b"` — binary mode (e.g., `"rb"`, `"wb"`)


## Writing to Files

Use `open("filename", "w")` with `file.write()` to write content.
**Warning:** `"w"` mode overwrites the file if it already exists!


```python
msg="Ahmad here !"
with open("introduction.txt","w") as file:
    file.write(msg)
    print(msg)
```

```python
with open("introduction.txt","r") as file:
    print(file.read())
```

## Reading, Modifying & Appending

Read a file, process the content in memory, then append the result.
- `file.read()` — reads entire file as a single string
- `file.readline()` — reads one line at a time
- `file.readlines()` — returns a list of lines
- `.replace()` is a string method, not a file method


```python
readedFile = "Usman"
with open("introduction.txt", "r") as readedFile:
    FetchedData = readedFile.read()
    print(FetchedData)
    FetchedData = FetchedData.replace("Ali","Ahmad")
    FetchedData = FetchedData.replace("50","25")
    FetchedData = FetchedData.replace("Data Analysis","Advance Python")
    print(FetchedData)
    with open("introduction.txt","a") as File2Write:
        File2Write.write(f"\n {FetchedData}")
```

## JSON File Handling

JSON is a lightweight data format. Python's `json` module handles it natively.
- `json.dump(obj, file)` — write object to file
- `json.load(file)` — read object from file


```python
import json

# Write to JSON
user_record = {
    "username": "ahmad",
    "email": "ahmad@example.com",
    "active": True,
    "scores": [85, 92, 78]
}

with open("user.json", "w") as jf:
    json.dump(user_record, jf, indent=4)
print("✅ user.json created!")
```

```python
# Read from JSON
with open("user.json", "r") as jf:
    loaded_data = json.load(jf)
    print("Loaded:", loaded_data)
    print("Username:", loaded_data["username"])
    print("Scores:", loaded_data["scores"])
```

## CSV File Handling

Use Python's `csv` module for comma-separated value files.
- `csv.writer(file)` — create a CSV writer
- `csv.reader(file)` — create a CSV reader
- `writer.writerow(list)` — write one row


```python
import csv

# Write a CSV file
with open("students.csv", "w", newline="") as cf:
    writer = csv.writer(cf)
    writer.writerow(["Name", "Age", "Grade"])  # header
    writer.writerow(["Ahmad", 25, "A"])
    writer.writerow(["Ali", 27, "B"])
    writer.writerow(["Sara", 22, "A+"])
print("✅ students.csv created!")
```

```python
# Read a CSV file
with open("students.csv", "r") as cf:
    reader = csv.reader(cf)
    for row in reader:
        print(row)
```

## Error Handling with Files

Always use `try/except` when working with files to handle missing files, permission errors, etc.


```python
try:
    with open("nonexistent.txt", "r") as f:
        print(f.read())
except FileNotFoundError:
    print("❌ File not found! Please check the filename.")
except PermissionError:
    print("❌ Permission denied! Cannot read the file.")
finally:
    print("Done with file operation.")
```

## OS Module for File Operations

The `os` module provides functions for interacting with the operating system.


```python
import os

filename = "introduction.txt"
print(f"'{filename}' exists:", os.path.exists(filename))
print("Current directory:", os.getcwd())
print("\nFiles in current directory:")
for f in os.listdir("."):
    if f.endswith(".txt") or f.endswith(".csv"):
        print(f"  📄 {f}")
```
