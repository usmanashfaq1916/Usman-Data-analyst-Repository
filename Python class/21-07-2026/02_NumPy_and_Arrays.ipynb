Numpy Again


````python
import time
import numpy as np

# 1. Setup data size (10 million items)
DATA_SIZE = 10_000_000

# Initialize a standard Python list and a NumPy array
py_list = list(range(DATA_SIZE))
np_array = np.arange(DATA_SIZE)

print(f"Benchmarking operations over {DATA_SIZE:,} elements...\n")

# --- TEST 1: SUMMATION ---
# Measure Python list sum
start_time = time.time()
list_sum = sum(py_list)
list_sum_time = time.time() - start_time

# Measure NumPy array sum
start_time = time.time()
array_sum = np.sum(np_array)
array_sum_time = time.time() - start_time

print("### Test 1: Summing Elements ###")
print(f"Python List time: {list_sum_time:.5f} seconds")
print(f"NumPy Array time: {array_sum_time:.5f} seconds")
print(f"👉 NumPy is {list_sum_time / array_sum_time:.1f}x faster\n")


# --- TEST 2: ELEMENT-WISE MULTIPLICATION ---
# Measure Python list multiplication (requires a loop/list comprehension)
start_time = time.time()
list_mult = [x * 2 for x in py_list]
list_mult_time = time.time() - start_time

# Measure NumPy array multiplication (uses optimized vectorization)
start_time = time.time()
array_mult = np_array * 2
array_mult_time = time.time() - start_time

print("### Test 2: Multiplying Elements by 2 ###")
print(f"Python List time: {list_mult_time:.5f} seconds")
print(f"NumPy Array time: {array_mult_time:.5f} seconds")
print(f"👉 NumPy is {list_mult_time / array_mult_time:.1f}x faster\n")
````

## Array Operations & Broadcasting

NumPy performs element-wise operations without loops (vectorization). Broadcasting automatically expands smaller arrays to match larger ones.


````python
a = np.array([1, 2, 3, 4])
b = np.array([10, 20, 30, 40])

print(a + b)    # [11 22 33 44] — element-wise addition
print(a * b)    # [ 10  40  90 160] — element-wise multiplication
print(a ** 2)   # [ 1  4  9 16] — square each element
print(a * 10)   # [10 20 30 40] — broadcasting: scalar × array
````

````python
studentMarks = np.array([25 , 50, 80 , 90 ,10 , 33])
# 80 > A
# 80 < B
# 50 < C
# 33 <= F 
````

````python
gradeList = []
for i in studentMarks:
    if i >=80:
        gradeList.append("A")
    elif i<80 and i >= 50:
        gradeList.append("B")
    elif i>= 33 and i < 50:
        gradeList.append("C")
    elif i<=33:
        gradeList.append("F")
gradeArray = np.array(gradeList)
````

````python
print(studentMarks)
print(gradeArray)
````

````python
studentMarks = np.array([25 , 50, 80 , 90 ,10 , 33])

condition = [(studentMarks >=80),(studentMarks >=50),(studentMarks>=33)]
grade = ["A","B","C"]
gradeArray = np.select(condition,grade,default="F")
print(gradeArray)
````

## Statistical Functions

NumPy provides many statistical functions that operate on entire arrays:
- `np.mean(arr)`, `np.median(arr)`, `np.std(arr)`
- `np.min(arr)`, `np.max(arr)`, `np.sum(arr)`
- Use `axis=0` for columns, `axis=1` for rows


````python
data = np.array([12, 45, 23, 67, 34, 89, 15, 42])
print(f"Mean:   {np.mean(data):.2f}")
print(f"Median: {np.median(data):.2f}")
print(f"Std:    {np.std(data):.2f}")
print(f"Min:    {np.min(data)}")
print(f"Max:    {np.max(data)}")
print(f"Sum:    {np.sum(data)}")

# Axis example with 2D array
matrix = np.array([[1, 2, 3], [4, 5, 6]])
print("Row sums:", np.sum(matrix, axis=1))    # [ 6 15]
print("Col sums:", np.sum(matrix, axis=0))    # [5 7 9]
````

## Random Module

`np.random` generates random numbers for simulations and testing:
- `np.random.randint(low, high, size)` — random integers
- `np.random.rand(d0, d1, ...)` — uniform [0, 1)
- `np.random.randn(d0, d1, ...)` — standard normal distribution


````python
np.random.seed(42)  # for reproducible results

rand_ints = np.random.randint(1, 101, 10)
print("Random integers:", rand_ints)

rand_3x3 = np.random.rand(3, 3)
print("Random 3x3 matrix:\n", rand_3x3)

normal_data = np.random.randn(1000)
print(f"Mean: {np.mean(normal_data):.4f}, Std: {np.std(normal_data):.4f}")
````
