# Pandas Data Analysis

Pandas is the most popular Python library for data analysis and manipulation. It provides two main data structures:
- **Series** — a 1D labeled array (like a column in a spreadsheet)
- **DataFrame** — a 2D labeled table (like a spreadsheet or SQL table)

Pandas can read/write Excel, CSV, JSON, SQL databases, and more.

## Row Filtering & Selection

Filter rows using boolean conditions:
- `df[df["column"] > value]` — conditional filtering
- `df.loc[row_label, col_label]` — label-based selection
- `df.iloc[row_index, col_index]` — integer position-based selection


```python
# Filtering example with the student DataFrame
print("Students older than 20:")
print(data[data["age"] > 20])
print("\nNames of students older than 20:")
print(data[data["age"] > 20]["Name"])
```

## Handling Missing Data

- `.dropna()` — remove rows with NaN
- `.fillna(value)` — fill NaN with a value
- `.isna()` / `.notna()` — detect missing values


```python
# Handling missing values
itemsDf.isna().sum()  # count NaN in each column
```

```python
#Store data of atleast 5 students with FName , Lname ,age , Class and gender

# Build banking and use csv/excel as db
```

```python
df = pd.read_csv("bank.csv")
bank = {}
for i in range(len(df)):
    account = str(df["AccountNumber"][i])
    bank[account] = {
        "FName": df["FName"][i],
        "LName": df["LName"][i],
        "Password": str(df["Password"][i]),
        "Address": df["Address"][i],
        "Class": df["Class"][i],
        "City": df["City"][i],
        "Balance": float(df["Balance"][i])
    }
while True:
    print("\n1. Create Account")
    print("2. Login")
    print("3. Exit")
    choice = input("Enter Choice: ")
    if choice == "1":
        account = input("Account Number: ")
        if account in bank:
            print("Account Already Exists")
        else:
            bank[account] = {}
            bank[account]["FName"] = input("First Name: ")
            bank[account]["LName"] = input("Last Name: ")
            bank[account]["Password"] = input("Password: ")
            bank[account]["Address"] = input("Address: ")
            bank[account]["Class"] = input("Class: ")
            bank[account]["City"] = input("City: ")
            bank[account]["Balance"] = float(input("Opening Balance: "))
            print("Account Created")

    elif choice == "2":
        account = input("Account Number: ")
        password = input("Password: ")
        if account in bank and password == bank[account]["Password"]:
            while True:
                print("\n1. View Details")
                print("2. Check Balance")
                print("3. Deposit")
                print("4. Withdraw")
                print("5. Logout")
                option = input("Choice: ")
                if option == "1":
                    print("First Name :", bank[account]["FName"])
                    print("Last Name :", bank[account]["LName"])
                    print("Address :", bank[account]["Address"])
                    print("Class :", bank[account]["Class"])
                    print("City :", bank[account]["City"])
                    print("Balance :", bank[account]["Balance"])
                elif option == "2":
                    print("Balance :", bank[account]["Balance"])
                elif option == "3":
                    amount = float(input("Amount: "))
                    bank[account]["Balance"] = bank[account]["Balance"] + amount
                    print("New Balance :", bank[account]["Balance"])
                elif option == "4":
                    amount = float(input("Amount: "))
                    if amount <= bank[account]["Balance"]:
                        bank[account]["Balance"] = bank[account]["Balance"] - amount
                        print("Remaining Balance :", bank[account]["Balance"])
                    else:
                        print("Insufficient Balance")
                elif option == "5":
                    break
                else:
                    print("Invalid Choice")
        else:
            print("Invalid Account or Password")
    elif choice == "3":
        accounts = []
        fname = []
        lname = []
        password = []
        address = []
        student_class = []
        city = []
        balance = []
        for account in bank:
            accounts.append(account)
            fname.append(bank[account]["FName"])
            lname.append(bank[account]["LName"])
            password.append(bank[account]["Password"])
            address.append(bank[account]["Address"])
            student_class.append(bank[account]["Class"])
            city.append(bank[account]["City"])
            balance.append(bank[account]["Balance"])
        df = pd.DataFrame({
            "AccountNumber": accounts,
            "FName": fname,
            "LName": lname,
            "Password": password,
            "Address": address,
            "Class": student_class,
            "City": city,
            "Balance": balance
        })
        df.to_csv("bank.csv", index=False)
        print("Data Saved")
        break
    else:
        print("Invalid Choice")
```
