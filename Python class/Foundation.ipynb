```python
number1 = int(input("Enter first number: "))
number2 = int(input("Enter second number: "))
operator = input("Enter operator (+,-,* /): ")
if operator == "+":
    result = number1 + number2
    if result % 2 == 0:
        print(f"Result {result} and it is even number")
    elif result % 2 != 0:
        print(f"Result {result} and it is odd number")
elif operator == "-":
    result = number1 - number2
    if result % 2 == 0:
        print(f"Result {result} and it is even number")
    elif result % 2 != 0:
        print(f"Result {result} and it is odd number")
elif operator == "*":
    result = number1 * number2
    if result % 2 == 0:
        print(f"Result {result} and it is even number")
    elif result % 2 != 0:
        print(f"Result {result} and it is odd number")
elif operator == "/":
    result = number1 / number2
    if result % 2 == 0:
        print(f"Result {result} and it is even number")
    elif result % 2 != 0:
        print(f"Result {result} and it is odd number")
else:
    print("invalid entry")

```
