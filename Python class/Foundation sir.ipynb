```python
#Talke Two numbers from user and take input from user which operation to perfrom
# Check if results is even or odd and print the result
n1 = int(input("Enter Number 1: "))
n2 = int(input("Enter Number 2: "))
operation = input("Enter operation (+, -, *, /): ")

if operation == "+":
    result = n1 + n2
    if result % 2 == 0:
        print(f"The result of {n1} + {n2} = {result} and it is even")
    else:
        print(f"The result of {n1} + {n2} = {result} and it is odd")
elif operation == "-":
    result = n1 - n2    
    if result % 2 == 0:
        print(f"The result of {n1} - {n2} = {result} and it is even")
    else:
        print(f"The result of {n1} - {n2} = {result} and it is odd")

elif operation == "*":
    result = n1 * n2
    if result % 2 == 0:
        print(f"The result of {n1} * {n2} = {result} and it is even")
    else:
        print(f"The result of {n1} * {n2} = {result} and it is odd")
elif operation == "/":
    result = n1 / n2
    if int(result) % 2 == 0:
        print(f"The result of {n1} / {n2} = {result} and it is even")
    else:
        print(f"The result of {n1} / {n2} = {result} and it is odd")
else:
    print("Invalid operation")
    result = None
```

```python
"""
Problem 5: The URL Parser
Goal: You are building a tool that extracts the main domain name from a simple 
secure website URL.

The input URL will always start with "https://www." and end with .com.

Write a program that uses slicing to cut off the "https://www." from the front and 
the .com from the back, t
hen print just the domain name.

Hint: Count the characters in "https://www." to know where your slice should start. 
Use negative indexing to cut off the .com.
"""
```

```python
url = "https://www.example.com"
if url[0:12]== "https://www." and url[-4:] == ".com":
    domain_name = url[12:-4]
    print(domain_name)
if url.startswith("https://www.") and url.endswith(".com"):
    domain_name = url[12:-4]
    print(domain_name)
```
